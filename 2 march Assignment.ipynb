Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np

A scatter plot is a simple graphical representation of data points in a two-dimensional coordinate system. It uses individual data points to show the relationship between two variables. Each data point is represented as a dot, and the position of the dot on the graph is determined by the values of the two variables being compared. This visualization helps to identify patterns, trends, and correlations between the variables.

```python
import matplotlib.pyplot as plt 
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("This is x Axis ")
plt.ylabel("This is Y- Axis")
plt.title("This is Title")
plt.scatter(x,y)
```

The subplot() function is used in various programming environments and libraries (e.g., Matplotlib in Python) to create multiple plots within a single figure or canvas. It allows you to divide the plotting area into a grid of rows and columns, and then you can create individual plots (subplots) within each grid cell.

```python
import numpy as np
import matplotlib.pyplot as plt 
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])  
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])  
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])  
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x1,y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x2,y2)
plt.title("Line2")

plt.subplot(2, 2, 3)
plt.plot(x3,y3)
plt.title("Line3")

plt.subplot(2, 2, 4)
plt.plot(x4,y4)
plt.title("Line4")\

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a graphical representation that uses rectangular bars to show the values of different categories or groups. The length or height of each bar is proportional to the data it represents. Bar plots are commonly used to compare and visualize the values of different categories or groups, making it easy to identify patterns, trends, and disparities between them. They are particularly useful when dealing with discrete data or when presenting categorical information, making it easier to understand and interpret the data at a glance.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()
plt.barh(company,profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot is a statistical visualization that displays the distribution of data through quartiles. It consists of a rectangular "box" representing the interquartile range (IQR), a line inside the box indicating the median, and "whiskers" extending from the box to show the range of the data.
It is used to summarize and compare the distribution of numerical data, providing a quick overview of key statistics such as the median, spread, and outliers. 

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
Box= [box1, box2]
plt.boxplot(Box)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for box plots
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a list
data = [box1, box2]

# Create the box plot
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.ylabel('Values')
plt.grid(True)

plt.show()
```
