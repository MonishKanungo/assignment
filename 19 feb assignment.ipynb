#Q1 What is Flask Framework? What are the advantages of Flask Framework?

Flask is a microframework for Python. It is designed to be simple and flexible, while providing a powerful set of features.

Advantages of Flask Framework:

Simple and easy to learn: Flask is a very simple framework, which makes it easy to learn and get started with. The documentation is also very good, which makes it easy to find help when you need it.
Flexible: Flask is a microframework, which means that you can choose which features and libraries you need. This gives you a lot of flexibility in how you build your application.
Extensible: Flask has a large and active community of developers who have created a wide range of extensions that add new features and functionality to Flask.
Secure: Flask is built on top of secure libraries and has a large and active community of developers who contribute to its security.

from flask import Flask
default=Flask(__name__)

@default.route("/")
def test():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    default.run(host="0.0.0.0")

#Q3. What is App routing in Flask? Why do we use app routes?


App routing in Flask is the process of mapping specific URLs to functions that will handle the logic for that URL. This allows you to create a web application that is easy to use and navigate.

App routes are used in Flask to:

Control access to different parts of your application
Group related functionality together
Make your application more modular

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

The url_for() function is used in Flask for URL building. It takes the name of the function as its first argument and any number of keyword arguments, each corresponding to a variable part of the URL rule. Unknown variable parts are appended to the URL as query parameters.
